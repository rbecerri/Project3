```python
# Labels are the values we want to predict
labels = np.array(data['HasDetections'])
#labels
```

```python
# Remove the labels from the features
# axis 1 refers to the columns
features = data.drop(columns =  ["MachineIdentifier", "HasDetections", "AvSigVersion", "OsBuildLab", 
                       "Census_MDC2FormFactor", "Census_DeviceFamily", "Census_ProcessorClass", "Census_PrimaryDiskTypeName", "Census_ChassisTypeName", 
                        "Census_PowerPlatformRoleName", "Census_InternalBatteryType", "Census_OSVersion", "Census_OSArchitecture", "Census_OSBranch", "Census_OSEdition", "Census_OSSkuName", 
                        "Census_OSInstallTypeName", "Census_OSWUAutoUpdateOptionsName", "Census_GenuineStateName", "Census_ActivationChannel", "Census_FlightRing"], axis = 1)
                               
#features = pd.get_dummies(features[:100000], columns = list(features.columns).remove('Census_SystemVolumeTotalCapacity'))
```

```python
#features2 = pd.get_dummies(features, columns = list(features.columns))

for column in features.columns:
    print(column)
    print(features.shape)
    features = pd.get_dummies(features, columns=[column], drop_first=True)
    print(features.shape)
```

```python
labels = labels[:100000]
```

```python
# Using Skicit-learn to split data into training and testing sets
from sklearn.model_selection import train_test_split
# Split the data into training and testing sets
train_features, test_features, train_labels, test_labels = train_test_split(features, labels, test_size = 0.25, random_state = 1)
```

```python
## Import the random forest model.

from sklearn.ensemble import RandomForestClassifier 
## This line instantiates the model. 
rf = RandomForestClassifier() 
## Fit the model on your training data.
rf.fit(train_features, train_labels)
```

```python
#Test Score
rf.score(test_features, test_labels)
```

```python
feature_importances = pd.DataFrame(rf.feature_importances_,
                                   index = train_features.columns,
                                    columns=['importance']).sort_values('importance', ascending=False)
feature_importances
```

```python
top_features = list(feature_importances[feature_importances['importance'] != 0].index)
```

```python
feature_importances[feature_importances['importance'] != 0].to_csv('feature_importances.csv')
```

```python
from sklearn.linear_model import LogisticRegression
classifier = LogisticRegression()

classifier.fit(train_features[top_features], train_labels)
```

```python
classifier.score(test_features[top_features],test_labels)
```
